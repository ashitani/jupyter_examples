# Slide Sample

iPython notebookによるスライドのテスト。

基本的なjupyterの使い方です。

matplotlib/numpyとセットで使うことが多いでしょう。

```python
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
t=np.linspace(0,10,100)
y=np.sin(t)

_=plt.plot(t,y)
```

画像を読み込んで表示します。この例は白黒だけど三色そろっているjpgです。つまりR=G=Bなもの。

```python
from PIL import Image
img = np.asarray(Image.open("sample.jpg"))
_=plt.imshow(img)
```

１チャネルしかない画像の場合はヒートマップ表示されます。

```python
img = np.asarray(Image.open("sample.jpg").convert("L")) #単色に変換
_=plt.imshow(img)
```

ヒートマップはヒートマップで便利な場合があります。

カラーバーを表示するとどの辺がどのぐらいの輝度なのかわかりやすいですね。

```python
img = np.asarray(Image.open("sample.jpg").convert("L")) #単色に変換
_=plt.imshow(img); _=plt.colorbar()
```

普通の見た目にしたいときはcmap="gray"を指定します。climを指定しないと輝度の最小から最大までをカラーマップに割り当てるので注意です。

```python
img = np.asarray(Image.open("sample.jpg").convert("L")) #単色に変換
_=plt.imshow(img,cmap="gray")
_=plt.clim(0,255)
_=plt.colorbar()
```
