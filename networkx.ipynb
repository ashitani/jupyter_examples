```python
import networkx as nx
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
vector={}
vector["start"]=["U","R"]
vector["U"]=["UR","UU"]
vector["R"]=["RR","RU"]

g = nx.DiGraph(vector)
nx.draw_networkx(g,pos=nx.graphviz_layout(g))
```

```python
g2 = nx.DiGraph({})
g2.add_nodes_from( ["a","b","c"])
g2.add_edge("a","b",{"name":"ab"})
g2.add_edge("a","c",{"name":"ac"})

edge_labels=dict([((u,v,),d['name']) for u,v,d in g2.edges(data=True)])

pos=nx.graphviz_layout(g2)
nx.draw_networkx(g2,pos=pos)
nx.draw_networkx_edge_labels(g2,pos=pos,edge_labels=edge_labels)
```
